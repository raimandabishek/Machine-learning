```python
# prompt: how to change the string columns into the interger

# Identify the columns that are strings and should be converted to integers
string_columns_to_convert = ['Name', 'Team', 'Position', 'College'] # Assuming these are the string columns you want to try to convert if they contain numbers

# Attempt to convert the identified string columns to numeric, coercing errors to NaN
for col in string_columns_to_convert:
    # Check if the column exists in the DataFrame before attempting conversion
    if col in df1.columns:
        df1[col] = pd.to_numeric(df1[col], errors='coerce')

# You can then check the info again to see the data types after conversion
df1.info()
```
